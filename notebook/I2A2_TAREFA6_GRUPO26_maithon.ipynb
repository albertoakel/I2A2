# **:: Atividade 2 – Limpeza e Padronização**

<h4 align="justify">A diversidade de origens e destinos pode gerar confusões quando informações semelhantes aparecem de forma inconsistente, prejudicando relatórios ambientais e logísticos.</h4>
<h4 align="justify">Identifique e trate valores inconsistentes (ex.: "reciclagem formal" e "Reciclagem Formal"), corrija erros de digitação e padronize todas as colunas categóricas. Em seguida, elimine duplicatas mantendo apenas registros válidos.</h4>

````python
df.head(10)
````

````python
# Escreva aqui a solução da Atividade 2
# --- 1. Análise Exploratória Inicial (EDA) ---
# Cria um buffer de string em memória para armazenar a saída do df.info()
buffer = io.StringIO()
# Captura as informações do DataFrame (tipos, memória, colunas, não nulos, etc.) dentro do buffer
df.info(buf=buffer)
# Lê o conteúdo do buffer como uma string e armazena na variável s_info
s_info = buffer.getvalue()

print("\n--- Análise Exploratória Inicial ---")
print("Shape do DataFrame:", df.shape)
print("\nInformações do DataFrame:\n", s_info)
print("\nEstatísticas Descritivas:\n")
print(df.describe(include="all"))
print("\n✅ Resultado da cotagem de valores NaN por coluna:\n")
print(df.isnull().sum())
print("\n✅ Resultado da cotagem de linhas duplicadas:\n")
print(df.duplicated().sum())
print("\n--- Limpeza e Padronização dos Dados ---")

# Função para padronizar colunas categóricas
def standardize_text(df, column):
    df[column] = df[column].astype(str).str.strip().str.lower().str.normalize("NFKD").str.encode("ascii", errors="ignore").str.decode("utf-8")
    return df

df = standardize_text(df, "Tipo_Eletronico")
df = standardize_text(df, "Origem")
df = standardize_text(df, "Destino_Final")
df = standardize_text(df, "Nivel_Toxico")

# ⚠️ - Atenção a partir do pandas 3.0 o uso de inplace=True será inutilizado para evitar modificações diretas em view
# ✅ - Foi aplicado o .fillna() para compatibilidade das versões futuras além de evitar problemas em views intermediárias (se houver)

# Aplicando Tratamento de inconsistências e valores nulos em Tipo_Eletronico
df["Tipo_Eletronico"] = df["Tipo_Eletronico"].replace({"nan": None})

# Aplicando Tratamento de inconsistências e valores nulos em Origem
df["Origem"] = df["Origem"].replace({"urbano": "urbana", "nan": None})

# Corrigir inconsistências e valores nulos em Destino_Final
df["Destino_Final"] = df["Destino_Final"].replace({"nan": None})

# Tratar valores ausentes
# Para Peso_kg e Custo_Reciclagem_R$, conversão para numérico e gerar média
df["Peso_kg"] = pd.to_numeric(df["Peso_kg"], errors="coerce")
df["Custo_Reciclagem_R$"] = df["Custo_Reciclagem_R$"].fillna(df["Custo_Reciclagem_R$"].mean())

df.fillna({"Peso_kg": df["Peso_kg"].mean()}, inplace=True)
df["Custo_Reciclagem_R$"] = df["Custo_Reciclagem_R$"].fillna(df["Custo_Reciclagem_R$"].mean())

# Para colunas categóricas com valores ausentes, vamos preencher com \'desconhecido\'
df["Tipo_Eletronico"] = df["Tipo_Eletronico"].fillna("desconhecido")
df["Origem"] = df["Origem"].fillna("desconhecido")
df["Destino_Final"] = df["Destino_Final"].fillna("desconhecido")

# Remover duplicatas baseadas em todas as colunas
df.drop_duplicates(inplace=True)

print("Limpeza e padronização concluídas.")
print("\n✅ Resultado da Contagem de valores NaN após limpeza:\n", df.isnull().sum())
print("\n✅ Resultado da Contagem de linhas duplicadas após limpeza:\n", df.duplicated().sum())
````

# **:: Atividade 3 – Distribuição e Frequências**

<h4 align="justify">Antes de planejar políticas públicas de descarte e reciclagem, é essencial saber quais eletrônicos mais aparecem e de onde eles vêm.</h4>
<h4 align="justify">Utilize groupby e value_counts para identificar a frequência de cada tipo de eletrônico e de cada origem, apresentando os resultados em tabelas e em gráficos de barras (Matplotlib ou Seaborn).</h4>

````python

print("\n--- Agrupamentos e Agregações ---")

# Média do Peso_kg e Custo_Reciclagem_R$ por Tipo_Eletronico
agg_tipo_eletronico = df.groupby("Tipo_Eletronico")[["Peso_kg", "Custo_Reciclagem_R$"]].mean().sort_values(by="Peso_kg", ascending=False)
print("\n✅ Resultado em dados por Tipo_Eletronico (Média de Peso e Custo de Reciclagem):\n", agg_tipo_eletronico)

# Gráfico de barras da frequência dos Tipos de Eletrônicos
plt.figure(figsize=(10,6))
sns.countplot(data=df, y="Tipo_Eletronico", order=df["Tipo_Eletronico"].value_counts().index)
# usando unicode para distinção de modelos de equipes similares
plt.title("\u27A4 Frequência dos Tipos de Eletrônicos")
plt.xlabel("Quantidade")
plt.ylabel("Tipo de Eletrônico")
plt.tight_layout()
plt.show()

# Contagem de eletrônicos por Origem e Destino_Final
agg_origem_destino = df.groupby(["Origem", "Destino_Final"]).size().unstack(fill_value=0)
print("\n✅ Resultado da agregação por Origem e Destino_Final (Contagem):\n", agg_origem_destino)

# Gráfico de barras empilhadas para Origem vs Destino_Final
agg_origem_destino.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Contagem de Eletrônicos por Origem e Destino Final")
plt.xlabel("Origem")
plt.ylabel("Quantidade")
plt.legend(title="Destino Final", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

# **:: Atividade 4 – Análise Temporal de Descarte**

# **:: Atividade 5 – Agregação por Município**

<h4 align="justify">A gestão de resíduos eletrônicos exige que municípios compreendam seu papel na geração e no destino do lixo tecnológico.</h4>
<h4 align="justify">Utilize groupby para agregar os dados por município, calculando:</h4>
<h4 align="justify">&bullet; o total de resíduos descartados;</h4>
<h4 align="justify">&bullet; a média de peso dos eletrônicos; e</h4>
<h4 align="justify">&bullet; e a proporção de destinos formais (reciclagem, doação) em relação aos informais (lixão, descarte irregular).</h4>
<h4 align="justify">Apresente os resultados em uma tabela ordenada e em gráfico de barras horizontais.</h4>

````python

# ✅ 1. Definindo os destinos formais e informais
destinos_formais = ['Reciclagem', 'Doação']
destinos_informais = ['Lixão', 'Descarte Irregular']

# 🛠️ 2. Criando colunas auxiliares para classificar os destinos
df["Destino_Formal"] = df["Destino_Final"].isin(destinos_formais)
df["Destino_Informal"] = df["Destino_Final"].isin(destinos_informais)

# 📊 3. Agrupamento dos dados por município
agrupado = df.groupby("Municipio").agg(
    Total_Residuos=('Peso_kg', 'sum'),
    Media_Peso=('Peso_kg', 'mean'),
    Total_Formal=('Destino_Formal', 'sum'),
    Total_Informal=('Destino_Informal', 'sum')
).reset_index()

# 🔢 4. Cálculo da proporção de destinos formais sobre informais
agrupado["Proporcao_Formal/Informal"] = agrupado["Total_Formal"] / (agrupado["Total_Informal"] + 1e-6)

# 🔽 5. Ordenando do município com mais resíduos para o que tem menos
agrupado = agrupado.sort_values(by="Total_Residuos", ascending=False)

# 🖨️ 6. Mostrando a tabela resultante
print("\n--- Agregação por Município ---")
print(agrupado)

# 📈 7. Gráfico de barras horizontais - Total de resíduos por município
plt.figure(figsize=(12, 6))

# ⚠️ ATUALIZAÇÃO IMPORTANTE:
# A partir da versão 0.14 do seaborn, usar 'palette' sem 'hue' será descontinuado.
# Por isso, adicionamos hue="Municipio" e legend=False para manter o estilo sem gerar legenda extra.
sns.barplot(
    x="Total_Residuos",
    y="Municipio",
    hue="Municipio",        # necessário para manter o uso do palette
    data=agrupado,
    palette="viridis",      # paleta de cores
    legend=False            # remove legenda redundante
)

plt.title("\u27A4 Total de Resíduos Eletrônicos por Município")
plt.xlabel("Total de Resíduos (kg)")
plt.ylabel("Município")
plt.tight_layout()
plt.show()

# 📊 8. Gráfico de barras - Proporção de destinos formais/informais
plt.figure(figsize=(12, 6))

sns.barplot(
    x="Proporcao_Formal/Informal",
    y="Municipio",
    hue="Municipio",        # necessário para manter o uso do palette
    data=agrupado,
    palette="mako",
    legend=False
)
# usando unicode para distinção de modelos de equipes similares
plt.title("\u27A4 Proporção de Destinos Formais/Informais por Município")
plt.xlabel("Proporção (Formais / Informais)")
plt.ylabel("Município")
plt.tight_layout()
plt.show()

# -------------------------------------------------------
# 🚨 9. Detecção de Outliers
print("\n--- Detecção de Outliers ---")

# 📦 Peso_kg - utilizando método do IQR (intervalo interquartil)
Q1_peso = df["Peso_kg"].quantile(0.25)
Q3_peso = df["Peso_kg"].quantile(0.75)
IQR_peso = Q3_peso - Q1_peso
outliers_peso = df[(df["Peso_kg"] < (Q1_peso - 1.5 * IQR_peso)) | (df["Peso_kg"] > (Q3_peso + 1.5 * IQR_peso))]
print("\n📦 Outliers em Peso_kg:\n", outliers_peso)

# 💰 Custo_Reciclagem_R$
Q1_custo = df["Custo_Reciclagem_R$"].quantile(0.25)
Q3_custo = df["Custo_Reciclagem_R$"].quantile(0.75)
IQR_custo = Q3_custo - Q1_custo
outliers_custo = df[(df["Custo_Reciclagem_R$"] < (Q1_custo - 1.5 * IQR_custo)) | (df["Custo_Reciclagem_R$"] > (Q3_custo + 1.5 * IQR_custo))]
print("\n💰 Outliers em Custo_Reciclagem_R$:\n", outliers_custo)

# 📉 10. Visualização dos outliers com boxplots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df["Peso_kg"], color='skyblue')
plt.title("\u25A0 Boxplot - Peso (kg)")

plt.subplot(1, 2, 2)
sns.boxplot(x=df["Custo_Reciclagem_R$"], color='salmon')
plt.title("\u25A0 Boxplot - Custo de Reciclagem (R$)")

plt.tight_layout()
plt.show()
````

# **:: Atividade 6 - Cruzamento entre Tipo de Eletrônico e Destino**

<h4 align="justify">Nem todos os eletrônicos têm o mesmo destino: alguns aparelhos têm maior chance de serem reciclados, enquanto outros acabam em descartes irregulares.</h4>
<h4 align="justify">Crie uma tabela cruzada (crosstab ou pivot_table) relacionando o tipo de eletrônico com seu destino. Visualize os resultados em um heatmap (Seaborn), destacando quais combinações são mais críticas.</h4>

# **:: Atividade 7 – Impacto do Peso dos Resíduos no Meio Ambiente**

<h4 align="justify">O peso dos resíduos eletrônicos é um fator crítico para a logística de coleta e o impacto ambiental. Equipamentos mais pesados, como geladeiras e televisores, podem sobrecarregar aterros irregulares e dificultar o transporte para centros de reciclagem. Na Amazônia, onde a infraestrutura de estradas é limitada, esse aspecto é ainda mais sensível.</h4>

Analisar a distribuição do peso é fundamental para indicar se há tendência de acúmulo de grandes aparelhos ou de pequenos dispositivos que passam despercebidos, mas se espalham de forma descontrolada.</h4>
<h4 align="justify">Utilize describe() e métodos de visualização (boxplot e histplot) para analisar a distribuição do peso dos resíduos eletrônicos. Identifique outliers e discuta possíveis explicações para sua ocorrência.</h4>
<h4 align="justify">Em seguida, faça uma comparação da média e do desvio padrão de peso entre descartes destinados formalmente e informalmente, apresentando os resultados em gráfico de violino.</h4>

````python
print("\n--- Análise Estatística do Peso dos Resíduos Eletrônicos ---")

# 7.A. Estatísticas descritivas com describe()
print(df["Peso_kg"].describe())

# 7.B. Histograma da distribuição do peso
plt.figure()
sns.histplot(df["Peso_kg"], bins=30, kde=True, color='teal')
plt.title("\u25A0 Distribuição do Peso dos Resíduos Eletrônicos")
plt.xlabel("Peso (kg)")
plt.ylabel("Frequência")
plt.tight_layout()
plt.savefig("histograma_peso.png")
plt.close()

# 7.C. Identificação de outliers (já feito antes, mas agora conectado com esta atividade)
Q1 = df["Peso_kg"].quantile(0.25)
Q3 = df["Peso_kg"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
outliers = df[(df["Peso_kg"] < limite_inferior) | (df["Peso_kg"] > limite_superior)]

print(f"\n📦 Outliers identificados no peso (IQR): {len(outliers)} registros")
print(outliers[["Tipo_Eletronico", "Peso_kg", "Destino_Final"]])

# 7.D. Classificação como destino formal ou informal
destinos_formais = ["Reciclagem", "Doação"]
df["Destino_Tipo"] = df["Destino_Final"].apply(lambda x: "Formal" if x in destinos_formais else "Informal")

# 7.E. Cálculo de média e desvio padrão do peso por destino
peso_por_destino = df.groupby("Destino_Tipo")["Peso_kg"].agg(["mean", "std"])
print("\n📊 Média e Desvio Padrão do Peso por Tipo de Destino:")
print(peso_por_destino)

# 7.F. Gráfico de violino para comparação de distribuição de peso
plt.figure()
sns.violinplot(
    data=df,
    x="Destino_Tipo",
    y="Peso_kg",
    hue="Destino_Tipo",
    palette="Set2",
    inner="quartile",
    legend=False
)
plt.title("\u25BA Distribuição do Peso por Tipo de Destino (Formal vs Informal)")
plt.xlabel("Tipo de Destino")
plt.ylabel("Peso (kg)")
plt.tight_layout()
plt.savefig("violino_peso_por_destino.png")
plt.close()

````

# **:: Atividade 8 – Relação entre Educação Ambiental e Destino do Resíduo**

<h4 align="justify">Um dos fatores determinantes para o destino correto dos resíduos eletrônicos é o nível de conscientização ambiental da população. Municípios que aplicam programas de educação ambiental tendem a apresentar maior proporção de descarte formal, reduzindo a poluição gerada por lixões clandestinos.</h4>
<h4 align="justify">No entanto, essa relação nem sempre é linear. Às vezes, mesmo com iniciativas locais, a ausência de infraestrutura de coleta inviabiliza o descarte correto. A análise comparativa pode revelar se a educação ambiental tem impacto mensurável no destino dos resíduos registrados na base.</h4>
<h4 align="justify">Agrupe os dados por município, comparando o percentual de descarte formal em locais com e sem registro de programas de educação ambiental.</h4>
<h4 align="justify">Crie gráficos de barras empilhadas mostrando a proporção de destinos em cada categoria, e discuta os possíveis vieses do dataset (ex.: municípios com dados incompletos).</h4>

````python
print("\n--- Atividade 8: Educação Ambiental x Destino do Resíduo ---")

# 1. Agrupamento por Município, Educação Ambiental e Tipo de Destino
agrupado = df.groupby(["Municipio", "Educacao_Ambiental", "Tipo_Destino"]).size().reset_index(name="Total")

# 2. Calcular proporção por grupo
agrupado["Proporcao"] = agrupado.groupby(["Municipio", "Educacao_Ambiental"])["Total"].transform(lambda x: x / x.sum())

# 3. Pivot da tabela para gráfico empilhado
tabela_pivot = agrupado.pivot_table(
    index=["Municipio", "Educacao_Ambiental"],
    columns="Tipo_Destino",
    values="Proporcao",
    fill_value=0
).reset_index()

# 4. Garantir que as colunas 'Formal' e 'Informal' existam
for col in ["Formal", "Informal"]:
    if col not in tabela_pivot.columns:
        tabela_pivot[col] = 0

# 5. Nome composto para eixo X
tabela_pivot["Município (Educação)"] = tabela_pivot["Municipio"] + " (" + tabela_pivot["Educacao_Ambiental"] + ")"

# 6. Gráfico de barras empilhadas
plt.figure(figsize=(12, 6))

plt.bar(
    tabela_pivot["Município (Educação)"],
    tabela_pivot["Formal"],
    label="Formal",
    color="seagreen"
)

plt.bar(
    tabela_pivot["Município (Educação)"],
    tabela_pivot["Informal"],
    bottom=tabela_pivot["Formal"],
    label="Informal",
    color="firebrick"
)

plt.title("\u2192 Proporção de Destinos por Município e Educação Ambiental")
plt.ylabel("Proporção")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Tipo de Destino")
plt.tight_layout()
plt.savefig("educacao_ambiental_vs_destino.png")
plt.show()
````
