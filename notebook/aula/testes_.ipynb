Atividade 6 - Atualizando Dados de Campo
Durante a checagem, foi descoberto que um dos registros tinha a quantidade de indivíduos informada incorretamente.
Escolha um registro específico (ex: ID = "M010") e atualize o valor da coluna Qtd Indivíduos para um novo número (à sua escolha). Em seguida, exiba a linha alterada.

```python
# Exibindo os primeiros IDs disponíveis na base

print("Alguns IDs disponíveis para atualização:\n")
display(df.head())

id_alvo = input("\nDigite o ID do registro a ser corrigido (ex: M010): ")

if id_alvo in df['ID'].values:
    novo_qt = int(input("Digite o novo valor para 'Qtd Indivíduos': "))
    
    df.loc[df['ID']==id_alvo,'Qtd Indivíduos']=novo_qt
    print(f"\nRegistro do ID {id_alvo} atualizado com sucesso:")
    print(df[df['ID'] == id_alvo])
else:
    print(f"\nID '{id_alvo}' não encontrado na base de dados.")


    

# #df.loc[,'Qtd Indivíduos']

# #df[df['ID']==M001

# #df[df['ID']=='M001']
# df.loc[df['ID']=='M002','Qtd Indivíduos']=11
# df.loc[df['ID']=='M002','Qtd Indivíduos']
```
