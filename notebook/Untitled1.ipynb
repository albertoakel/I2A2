```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Análise descritiva básica
print("=== ANÁLISE INICIAL ===")
print(f"Período: {descartes_mensais['Data_Descarte'].min()} até {descartes_mensais['Data_Descarte'].max()}")
print(f"Total de meses analisados: {len(descartes_mensais)}")
print(f"Média mensal de descartes: {descartes_mensais['volume_descartes'].mean():.2f}")
print(f"Desvio padrão: {descartes_mensais['volume_descartes'].std():.2f}")

# 2. Extrair mês e ano para análise sazonal
descartes_mensais['Ano'] = descartes_mensais['Data_Descarte'].dt.year
descartes_mensais['Mes'] = descartes_mensais['Data_Descarte'].dt.month
descartes_mensais['Mes_Nome'] = descartes_mensais['Data_Descarte'].dt.month_name()

# 3. Análise por mês (agrupando todos os anos)
descarte_por_mes = descartes_mensais.groupby('Mes').agg(
    total_descartes=('volume_descartes', 'sum'),
    media_mensal=('volume_descartes', 'mean'),
    mediana_mensal=('volume_descartes', 'median'),
    desvio_padrao=('volume_descartes', 'std'),
    contagem=('volume_descartes', 'count')
).reset_index()

# Adicionar nome do mês
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 
         'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
descarte_por_mes['Mes_Nome'] = meses

print("\n=== TOP 5 MESES COM MAIOR MÉDIA DE DESCARTES ===")
top_meses = descarte_por_mes.nlargest(5, 'media_mensal')
print(top_meses[['Mes_Nome', 'media_mensal', 'mediana_mensal', 'desvio_padrao']])

print("\n=== COMPARAÇÃO MÊS COM MAIOR vs MENOR DESCARTE ===")
mes_maior = descarte_por_mes.loc[descarte_por_mes['media_mensal'].idxmax()]
mes_menor = descarte_por_mes.loc[descarte_por_mes['media_mensal'].idxmin()]

print(f"Mês com maior média: {mes_maior['Mes_Nome']} - {mes_maior['media_mensal']:.2f} descartes")
print(f"Mês com menor média: {mes_menor['Mes_Nome']} - {mes_menor['media_mensal']:.2f} descartes")
print(f"Diferença: {mes_maior['media_mensal'] - mes_menor['media_mensal']:.2f} descartes")

# 4. Teste estatístico para verificar sazonalidade
# ANOVA para comparar médias entre meses
grupos_meses = [descartes_mensais[descartes_mensais['Mes'] == mes]['volume_descartes'] 
                for mes in range(1, 13)]

f_stat, p_value = stats.f_oneway(*grupos_meses)
print(f"\n=== TESTE ESTATÍSTICO (ANOVA) ===")
print(f"F-statistic: {f_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ Há diferenças estatisticamente significativas entre os meses (p < 0.05)")
    print("Isso indica padrão sazonal nos descartes!")
else:
    print("❌ Não há diferenças estatisticamente significativas entre os meses")

# 5. Visualização dos resultados
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Gráfico 1: Média de descartes por mês
mes_order = ['January', 'February', 'March', 'April', 'May', 'June', 
             'July', 'August', 'September', 'October', 'November', 'December']
media_por_mes = descartes_mensais.groupby('Mes_Nome')['volume_descartes'].mean().reindex(mes_order)

bars = ax1.bar(media_por_mes.index, media_por_mes.values, 
               color=['red' if x == media_por_mes.max() else 
                      'lightblue' if x == media_por_mes.min() else 
                      'steelblue' for x in media_por_mes.values],
               alpha=0.8)

ax1.set_title('Média de Descartes por Mês (2015-2025)', fontsize=14, fontweight='bold')
ax1.set_xlabel('Mês')
ax1.set_ylabel('Média de Descartes')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True, alpha=0.3, axis='y')

# Adicionar valores nas barras
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{height:.1f}', ha='center', va='bottom', fontweight='bold')

# Gráfico 2: Evolução temporal com destaque sazonal
cores_meses = plt.cm.Set3(np.linspace(0, 1, 12))
for mes in range(1, 13):
    dados_mes = descartes_mensais[descartes_mensais['Mes'] == mes]
    ax2.plot(dados_mes['Data_Descarte'].values, dados_mes['volume_descartes'].values, 
             'o-', alpha=0.7, color=cores_meses[mes-1], label=meses[mes-1])

ax2.set_title('Evolução Temporal por Mês', fontsize=14, fontweight='bold')
ax2.set_xlabel('Ano')
ax2.set_ylabel('Volume de Descartes')
ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax2.grid(True, alpha=0.3)

# Gráfico 3: Boxplot por mês
dados_boxplot = []
nomes_meses = []
for mes in range(1, 13):
    dados_mes = descartes_mensais[descartes_mensais['Mes'] == mes]['volume_descartes']
    dados_boxplot.append(dados_mes)
    nomes_meses.append(meses[mes-1][:3])

ax3.boxplot(dados_boxplot, labels=nomes_meses)
ax3.set_title('Distribuição de Descartes por Mês', fontsize=14, fontweight='bold')
ax3.set_xlabel('Mês')
ax3.set_ylabel('Volume de Descartes')
ax3.grid(True, alpha=0.3)

# Gráfico 4: Heatmap sazonal
pivot_table = descartes_mensais.pivot_table(values='volume_descartes', 
                                           index='Ano', 
                                           columns='Mes', 
                                           aggfunc='mean')

sns.heatmap(pivot_table, annot=True, fmt='.1f', cmap='YlOrRd', ax=ax4)
ax4.set_title('Heatmap: Média de Descartes por Ano e Mês', fontsize=14, fontweight='bold')
ax4.set_xlabel('Mês')
ax4.set_ylabel('Ano')

plt.tight_layout()
plt.show()

# 6. Análise de tendência por mês específico
print("\n=== ANÁLISE DETALHADA DOS MESES COM MAIOR DESCARTE ===")
for index, row in top_meses.iterrows():
    mes_dados = descartes_mensais[descartes_mensais['Mes'] == row['Mes']]
    tendencia = np.polyfit(range(len(mes_dados)), mes_dados['volume_descartes'], 1)[0]
    
    print(f"\n{row['Mes_Nome']}:")
    print(f"  • Média: {row['media_mensal']:.2f}")
    print(f"  • Mediana: {row['mediana_mensal']:.2f}")
    print(f"  • Tendência anual: {'↑' if tendencia > 0 else '↓'} {abs(tendencia):.2f} por ano")
    print(f"  • Variação: {row['desvio_padrao']:.2f}")

# 7. Identificar outliers por mês
print("\n=== MESES COM MAIORES PICOS DE DESCARTE ===")
for mes in top_meses['Mes']:
    mes_dados = descartes_mensais[descartes_mensais['Mes'] == mes]
    max_descarte = mes_dados['volume_descartes'].max()
    ano_max = mes_dados.loc[mes_dados['volume_descartes'].idxmax(), 'Ano']
    
    print(f"{meses[mes-1]}: Pico de {max_descarte} descartes em {ano_max}")
```
